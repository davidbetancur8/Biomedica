```python
import pandas as pd
import dash
import dash_html_components as html
import dash_core_components as dcc
import plotly.graph_objects as go
import plotly.express as px
from jupyter_dash import JupyterDash
from dash.dependencies import Input, Output

from urllib.request import urlopen
import json
import numpy as np
import plotly
```

```python
data_dict = {
    "defunciones": {"link": "http://medata.gov.co/node/21745/download", "colname": "comuna_res"},
    "nacimientos": {"link": "http://medata.gov.co/node/21730/download", "colname": "comuna_res"},
    "intoxicaciones": {"link": "http://medata.gov.co/node/21766/download", "colname": "sivigila_datosbasicos.comuna"},
    "dengue": {"link": "http://medata.gov.co/node/21463/download", "colname": "sivigila_denguegrave.comuna"}
}
```

```python
df_defunciones = pd.read_csv(data_dict["defunciones"]["link"], sep=";")
```

```python
df_defunciones.head()
```

```python
comunas = df_defunciones[data_dict["defunciones"]["colname"]].unique()
comunas
```

```python
comunas_dict = {c[3:].lower().replace("corregimiento de ", ""): int(c[:2]) for c in comunas if c != "Sin informacion"}
comunas_dict
```

```python
df_tox = pd.read_csv(data_dict["intoxicaciones"]["link"], sep=";")
```

```python
df_tox.value_counts("sivigila_datosbasicos.comuna")
```

```python
bad_places = ['sin ubicacion en zona urbana', "sin informacion", "san sebastian de palmitas", 'no ubicada por site']
dfs_finales = []
for db in data_dict:
  cuentas = pd.read_csv(data_dict[db]["link"], sep=";").value_counts(data_dict[db]["colname"]).to_frame(db).reset_index()
  cuentas[data_dict[db]["colname"]] = cuentas[data_dict[db]["colname"]].str.lower().str.replace("corregimiento de ", "")
  if data_dict[db]["colname"] == "comuna_res":
    cuentas[data_dict[db]["colname"]] = cuentas[data_dict[db]["colname"]].apply(lambda x: f"Comuna {int(x[:2])}" if x != "sin informacion" else "Comuna 0")
  else:
    cuentas[data_dict[db]["colname"]] = cuentas[data_dict[db]["colname"]].apply(lambda x: f"Comuna {comunas_dict[x.lower()]}" if x.lower() not in bad_places else "Comuna 0")
    cuentas = cuentas.rename(columns={data_dict[db]["colname"]:"comuna_res"})
  
  dfs_finales.append(cuentas)
```

```python
df_final = dfs_finales[0]
for df in dfs_finales[1:]:
  df_final = df_final.merge(df, left_on="comuna_res", right_on="comuna_res")
df_final = df_final[df_final["comuna_res"] != "Comuna 0"]
```

```python
df_final
```

```python
with urlopen('https://raw.githubusercontent.com/davidbetancur8/geojsons/master/data/medellin.json') as response:
    geoJSON = json.load(response)

deptos = [dpto["properties"] for dpto in geoJSON["features"]]
df = pd.DataFrame(deptos).loc[:,["CODIGO", "NOMBRE", "IDENTIFICA"]]
```

```python
df_final = df.merge(df_final, left_on="IDENTIFICA", right_on="comuna_res")
df_final
```

```python
app = JupyterDash(__name__) 


app.layout = html.Div(id = 'parent', children = [
    html.H1(id = 'H1', children = 'Stocks :)', style = {'textAlign':'center',\
                                            'marginTop':40,'marginBottom':40}),

        dcc.Dropdown( id = 'dropdown',
        options = [
            {'label':'defunciones', 'value':'defunciones' },
            {'label': 'nacimientos', 'value':'nacimientos'},
            {'label': 'intoxicaciones', 'value':'intoxicaciones'},
            {'label': 'dengue', 'value':'dengue'},
            ],
        value = 'nacimientos'),
        dcc.Graph(id = 'bar_plot')
    ])
    
    
@app.callback(Output(component_id='bar_plot', component_property= 'figure'),
              [Input(component_id='dropdown', component_property= 'value')])
def graph_update(dropdown_value):
    print(dropdown_value)
    fig = px.choropleth(df_final, geojson=geoJSON, color=dropdown_value,
                        locations="NOMBRE", featureidkey="properties.NOMBRE", title="Natalidad en Medellín 2018-2020",
                        projection="mercator",scope="south america",color_continuous_scale=px.colors.sequential.Blues,
                        range_color = [df_final[dropdown_value].min(),df_final[dropdown_value].max()]
                    )
    fig.update_geos(fitbounds="locations", visible=False, showcountries=True, countrycolor="Black",
        showsubunits=True)
    fig.update_layout(
        font=dict(
            family="Courier New, monospace",
            size=25,
            color="#7f7f7f"
        )
    )
    fig.show()
    return fig 
```

```python
app.run_server(mode="inline", debug=True)
```
