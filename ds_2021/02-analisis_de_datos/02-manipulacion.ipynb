```python
import pandas as pd
```

```python
df_costumers = pd.read_csv("https://www.plus2net.com/python/download/customer.csv")
df_products = pd.read_csv("https://www.plus2net.com/python/download/products.csv")
df_sales = pd.read_csv("https://www.plus2net.com/python/download/sales.csv").drop("product", axis=1)

df_costumers = df_costumers.append(df_costumers.head(1))
df_costumers.iloc[-1, :] = [999, "David"]

df_sales = df_sales.append(df_sales.head(1))
df_sales.iloc[-1, :] = [10,	500,	3,	2,	"ABC"]
```

# Merge

<img src="https://storage.googleapis.com/gweb-cloudblog-publish/images/joins_1.max-1600x1600.png">

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.merge.html

## Inner (por defecto)

```python
df_sales.merge(df_costumers, left_on="c_id", right_on="c_id", how="inner")
```

## Left

```python
df_sales.merge(df_costumers, left_on="c_id", right_on="c_id", how="left")
```

## Right

```python
df_sales.merge(df_costumers, left_on="c_id", right_on="c_id", how="right")
```

```python
df_sales_costumers = df_sales.merge(df_costumers, left_on="c_id", right_on="c_id")
df_sales_costumers_products = df_sales_costumers.merge(df_products, left_on="p_id", right_on="p_id")

df_sales_costumers_products
```

# Groupby

```python
df_sales_costumers_products.groupby("Customer").groups
```

```python
df_sales_costumers_products.groupby("Customer").mean()
```

```python
df_sales_costumers_products.groupby("Customer").size()
```

```python
df_sales_costumers_products.groupby('Customer')['price'].agg(
    price_mean='mean',
    price_std='std'
)
```

```python
def mean_pesos(value):
    value_mean = value.mean()
    return value_mean * 4500
```

```python
df_sales_costumers_products.groupby('Customer')["price"].agg(mean_pesos)
```

```python
df_sales_costumers_products.groupby('Customer')["price"].agg(lambda x: x.mean()*4500)
```

```python
import requests
import json

nombre = "Raju"
url = f"https://api.nationalize.io/?name={nombre}"
json.loads(requests.get(url).text)["country"][0]["country_id"]
```

```python
def get_nal(nombre):
  url = f"https://api.nationalize.io/?name={nombre}"
  nal = json.loads(requests.get(url).text)["country"][0]["country_id"]
  return nal
```

```python
get_nal("David")
```

```python
df_sales_costumers_products["nacionality"] = df_sales_costumers_products["Customer"].apply(get_nal)
df_sales_costumers_products
```
