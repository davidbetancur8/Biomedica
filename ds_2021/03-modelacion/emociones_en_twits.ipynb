```python
import pandas as pd
import twint
import nest_asyncio
from tqdm import tqdm
nest_asyncio.apply()
```

```python
c = twint.Config()
c.Limit = 30
c.Username = "petrogustavo"
c.Filter_retweets = True
c.Store_csv = True
c.Output = f"petrogustavo.csv"
c.Hide_output = True
s = twint.run.Search(c)
```

```python
df = pd.read_csv("petrogustavo.csv")
df.loc[:5, ["date", "tweet"]]
```

```python
from pysentimiento import EmotionAnalyzer
```

```python
emotion_analyzer = EmotionAnalyzer(lang="es")
emotion_analyzer.predict("estoy muy feliz!!!")
```

```python
emotion_analyzer.predict("estoy muy feliz!!!").output
```

```python
df["emocion"] = df["tweet"].apply(lambda x: emotion_analyzer.predict(x).output)
```

```python
df.loc[:5, ["date", "tweet", "emocion"]]
```

```python
df["emocion"].value_counts()
```

```python
df[df["emocion"]=="anger"]["tweet"].values
```

```python
df[df["emocion"]=="joy"]["tweet"].values
```
